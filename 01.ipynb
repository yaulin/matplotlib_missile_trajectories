```python
class trajectory:
    data = pd.DataFrame()
    
    def __init__(self,f):
        self.data = pd.read_csv(f)
        
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
s1 = "1a.CSV"
f1 = dir1 + s1


t1 = trajectory(f1)

t = t1.data["TTIME_S"]

x = t1.data["TRXI_M"]/1000    #convert to km
y = t1.data["TRYI_M"]/1000    #convert to km
z = t1.data["TRZI_M"]/1000    #convert to km




ax = plt.figure().add_subplot(projection='3d')




ax.plot(x, y, z)
ax.set(xlabel = 'X, km', ylabel = 'Y, km', zlabel = 'Z, km')


plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(t, x)

ax.set(xlabel = 'Time', ylabel='X coordinate, km')
ax.grid()

plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(t, x)

ax.set(xlabel = 'Time', ylabel='Y coordinate, km')
ax.grid()

plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(t, z)

ax.set(xlabel = 'Time', ylabel='Z coordinate, km')
ax.grid()

plt.show()
```

```python
ground_range = t1.data['TGRNKM_KM']


fig, ax = plt.subplots()
ax.plot(t, ground_range)

ax.set(xlabel = 'Time', ylabel='Ground Range, km')
ax.grid()

plt.show()
```

```python
altitude = t1.data['TALTKM_KM']


fig, ax = plt.subplots()
ax.plot(t, altitude)

ax.set(xlabel = 'Time', ylabel='Altitude, km')
ax.grid()

plt.show()
```

```python
downrange_distance = t1.data['TDRNGE_M']/1000

fig, ax = plt.subplots()
ax.plot(t, downrange_distance)

ax.set(xlabel = 'Time', ylabel='Downrange distance, km')
ax.grid()

plt.show()
```

```python
crossrange_distance = t1.data['TCRNGE_M']/1000   # convert to km

fig, ax = plt.subplots()
ax.plot(t, crossrange_distance)

ax.set(xlabel = 'Time', ylabel='Crossrange distance, km')
ax.grid()

plt.show()
```

```python
earth_rel_velocity = t1.data['TVRMAG_M/S']

fig, ax = plt.subplots()
ax.plot(t, earth_rel_velocity)

ax.set(xlabel = 'Time', ylabel='Earth relative velocity, m/s')
ax.grid()

plt.show()
```
